```python

import scipy.stats as stats 
import numpy as np
import pandas as pd
import plotly.graph_objects as go

df= pd.read_csv('spendingscores.csv')
df
```

```python
df['Gender_new'] =  df.apply(lambda x: 1 if x['Gender'] == 'Male' else 0 if x['Gender'] == 'Female' else None, axis = 1)
df
```

```python
df['Gender'] = df['Gender_new']
df.drop(columns='Gender_new', axis = 1)
```

```python

import statsmodels.formula.api as smf

model = smf.ols(formula="SpendingScore ~ Gender + Age + AnnualIncome", data=df).fit()
print('summary', model.summary(),'\n')
print('Pvalues', model.pvalues,'\n')
print(model.bse, '\n')
print(model.params,'\n')
```

```python
import plotly.express as px

corr = df.corr()
fig = px.scatter_matrix(df, dimensions=['Age', 'Gender', 'AnnualIncome'],title='Correlation Matrix')
fig.show()
```
